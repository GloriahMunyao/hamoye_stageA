```python
#importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import webbrowser

# Loading the dataset from my local PC
file_path = 'C:\Hamoye\fbs.csv' 
df = pd.read_csv(r'C:\Hamoye\fbs.csv' , encoding='latin-1')

year_set = ['Y2014', 'Y2015', 'Y2016', 'Y2017', 'Y2018']
element_col = 'Element'

#calculating the average for each item
averageData = df.groupby(element_col)[year_set].mean()

# Plotting the bar graph
averageData.plot(kind='bar', figsize=(15, 6))
plt.xlabel(element_col)
plt.ylabel('Average ')
plt.title('Graph of Element and Average (2014-2018)')
plt.legend(title='Year')
plt.show()

# calculating the sum for each Unit
elementsum = df.groupby(element_col)[year_set].sum()
print(elementsum)
```

```python
year_set = ['Y2014', 'Y2015', 'Y2016', 'Y2017', 'Y2018']
country = 'Area'
year2014='Y2014'

#calculating the average for each Area for each year
averageData = df.groupby(country)[year_set].mean()

# Plotting the bar graph
averageData.plot(kind='bar', figsize=(10, 6))
plt.xlabel(country)
plt.ylabel('Average ')
plt.title('Graph of Area and Average (2014-2018)')
plt.legend(title='Year')
plt.show()

#calculating the sum for each Area for for each year
areasum = df.groupby(country)[year_set].sum()
print(areasum)



# Checking for null values 
#nullData = df['year2014'].isnull()
#print(nullData)


```
